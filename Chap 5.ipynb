```python
# p212 5-23

# MAIN FUNCTION:

def main():

    #get the amount of sales:
    sales=get_sales()
    
    #get the amount of advanced pay:
    advanced_pay=get_advanced_pay()
    
    #determine the commission rate:
    comm_rate=determine_comm_rate(sales)
    
    #calculate the pay:
    pay=sales*comm_rate - advanced_pay
    
    #display the amount of pay:
    print('The pay is $', format(pay,',.2f'),sep='')
    
    #determine whether the pay is negative:
    if pay<0:
        print('The Salesperson must reimburse')
        print('the company.')
    
# get_sales FUNCTION:

def get_sales():
    
    monthly_sales=float(input('Enter the monthly sales: '))
    
    return monthly_sales

# get_advanced_pay FUNCTION:

def get_advanced_pay():
    print('Enter the amount of advanced pay, or')
    print('enter 0 if no advanced pay was given.')
    advanced=float(input('Advanced pay: '))
    
    return advanced

# determine_comm_rate FUNCTION:

def determine_comm_rate(sales):
    # determine the commission rate:
    if sales<10000:
        rate=0.1
    elif sales>=10000 and sales <=14999.99:
        rate=0.12
    elif sales>=15000 and sales<=17999.99:
        rate=0.14
    elif sales>=18000 and sales<=21999.99:
        rate=0.16
        
    else:
        rate=0.18
        
    # Return the commission rate:
    return rate

main()
```

```python
# Returning Boolean Values: a binary variable that can have one of two possible values, 0 (false) or 1 (true): p215

# '%' is called the reminder operator.  It divides 2 numbers and returns the remainder of the division.  e.g. number%2 means
# if the reminder of number divides by 2 is equal to 0, then display a message indicating the number is even, or else display a 
# message indicating the number is odd i.e. 8/2=4.0 the reminder is 0; 5/2=2.5, the reminder is 0.5 = NOT 0 - see example below:

def is_even(number):
    
    if (number%2)==0:
        status = True
    else:
        status=False
        
    return status

# rewrite the if-else satement so it CALLS the is_even function determine whether number is even:

number = int(input('Enter a number: '))
if is_even(number):
    print('The number is even.')
else:
    print('The number is odd.')
```

```python
#p220  Storing Functions in MODULES

# a MODULE is a FILE that contains Python code.  Large programs are easier to debug and maintain wehn they are divided 
# into modules.

import math

# the circle module:
def area(radius):
    return math.pi * radius**2

def circumference(radius):
    return 2 * math.pi * radius

# the rectangle module:
def area(width, length):
    return width * length

def perimeter(width, length):
    return 2 * (width + length)
```

```python
# an example how we would import circle and rectangle module: p220
import circle
import rectangle

# Constants for the menu choices: 
AREA_CIRCLE_CHOICE = 1
CIRCUMFERENCE_CHOICE = 2
AREA_RECTANGLE_CHOICE = 3
PERIMETER_RECTANGLE_CHOICE = 4
QUIT_CHOICE = 5

# the main function:

def main():
    
    # the choice variable controls the loop and holds the user's menu choice:
    choice = 0
    
    while choice != QUIT_CHOICE:
        # display the menu:
        display_menu()
        
        # get teh user's choice: 
        choice = int(input('Enter your choice: '))
        
        # perform the selected action:
        if choice == AREA_CIRCLE_CHOICE:
            radius = float(input("Enter the circle's radius: "))
            print('The area is', circle.area(radius))
        elif choice == CIRCUMFERENCE_CHOICE:
            radius = float(input("Enter the circle's radius: "))
            print('The circumference is',circle.circumference(radius))
        elif choice == AREA_RECTANGLE_CHOICE:
            width = float(input("Enter the rectangle's width: "))
            length = float(input("Enter the rectangle's length: "))
            print('The area is', rectangle.area(width, length))
        elif choice == PERIMETER_RECTANGLE_CHOICE:
            width = float(input("Enter the rectangle's width: "))
            length = float(input("Enter the rectangle's length: "))
            print('The perimeter is',rectangle.perimeter(width, length))
        elif choice == QUIT_CHOICE:
            print('Exiting the program...')
            
        else: 
            print('Error: invalid selection.')
            
# the display_menu function displays a menu:

def display_menu():
    print(' MENU')
    print('1) Area of a circle')
    print('2) Circumference of a circle')
    print('3) Area of a rectangle')
    print('4) Perimeter of a rectangle')
    print('5) Quit')
    
main()
```
